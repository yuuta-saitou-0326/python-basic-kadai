```python
# Humanクラスを定義
class Human:
    # コンストラクタでnameとageを属性として持たせる
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 年齢に応じて大人かどうかを判定するメソッド
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}さんは大人です。")
        else:
            print(f"{self.name}さんは大人ではありません。")

# Humanインスタンスを複数生成してリストに追加
human_list = [
    Human("祐太", 38),
    Human("さくら", 18),
    Human("たろう", 20)
]

# リストの要素に対してcheck_adultメソッドを実行
for human in human_list:
    human.check_adult()
```
